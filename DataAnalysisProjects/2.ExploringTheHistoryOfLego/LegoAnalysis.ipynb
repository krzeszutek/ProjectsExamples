```python
data = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
display(data.head(20))
display(theme.head(20))
```

#### 1. What percentage of all licensed sets ever realeased were Star Wars themed?

*set_num* is a critical column. Mising value indicates that the set is a duplicate or invalid
  
  *set_name* - a name for every set in the dataset
  
  *year* - the date the set was released
  
  *num_parts* - The number of parts contained in the set - column is not central, so missing values are okay here

```python
# We need to join te data in one dataset
merged = data.merge(theme, left_on='parent_theme', right_on='name')
merged.head(20)

# Dropping the repeated columns
merged = merged.drop(columns='name_y')
```

```python
merged.head(20)
```

```python
# Checking how many null values are in the 
merged[merged['set_num'].isnull()].shape
```

```python
licensed = merged[merged['is_licensed'] == True]
licensed = licensed.dropna(subset=['set_num']) # We delete the values where set_num was zero
licensed['Count'] = 1
star_wars = licensed[licensed['parent_theme'] == 'Star Wars']
the_force = star_wars.shape[0]/licensed.shape[0]
```

```python
the_force
```

#### In which year was Star Wars not the most popular licensed theme (in terms of numbers of sets released that year?)

```python
licensed_sorted = licensed.sort_values('year')
summed_df = licensed_sorted.groupby(['year', 'parent_theme']).sum().reset_index()
summed_df.sort_values('Count', ascending=False).drop_duplicates(['year'])
```
