```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Generating random data
np.random.seed(0)
X = 2.5 * np.random.randn(100) + 1.5 # Generating 100 random data with mean = 1.65
res = 0.5 * np.random.randn(100) # Generating 100 residuals terms       
Y =  2 + 0.3 * X + res                  # Actual values of Y
print(Y)
```

```python
# Creating pandas data frame to store our X and Y values
df = pd.DataFrame({'X':X, 'Y': Y})
display(df.head(10))
```

```python
# Calculating the means
xmean = X.mean()
ymean = Y.mean()
df['covariance'] = (df['X'] - xmean) * (df['Y'] - ymean)
df['(xi - xmean)**2'] = (df['X'] - xmean)**2
```

```python
beta = sum(df['covariance'])/sum(df['(xi - xmean)**2'])
alpha = ymean - beta*xmean
print('alpha :', alpha)
print('beta:', beta)
```

```python
# We can make predictions with our model
ypred = alpha + beta * X
```

```python
# Plotting our predictions against actual data
plt.figure(figsize=(12,6))
plt.plot(X, ypred)
plt.plot(X, Y, 'ro')
plt.title('Actual vs Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
